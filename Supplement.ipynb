```python
# Comparison of different ATPase models
from plm_singlecomp_withkcc2 import plm, default_p, F, nao, qpump
import matplotlib.pyplot as plt
import numpy as np

vel0=[]
vel1=[]
vel2=[]

naspace=range(1,22)

for na in naspace:
    na=na*0.001 
    jp=10**(default_p)/(F)*(na/nao)**3 # cubic pump rate update (dependent on sodium gradient)
    vel0.append(jp)
    jp=(0.014/nao)**2*10**(default_p)/F*na/nao # linear
    vel1.append(jp)
    jp=qpump*(1.62/(1+(0.0067/na)**3)+1.0/(1+(0.0676/na)**3))/F # hamada
    vel2.append(jp)

plt.plot(naspace,vel0,'k')
plt.plot(naspace,vel1,'g')
plt.plot(naspace,vel2,'r')
plt.savefig('naspace.eps')
plt.show()
```

```python
# Comparison of different KCC2 models
from plm_singlecomp_withkcc2 import plm, vmax, kd, gkcc, cle, ke, F, R
import matplotlib.pyplot as plt
import numpy as np

pkcc=gkcc
k=0.118

vel0=[]
vel1=[]
vel2=[]
vel3=[]

clspace=range(1,54)

for cl in clspace:
    cl=cl*0.001 
    externals=cle*ke/k
    fix=np.log(externals/0.056)
    jkcc2=-0.117*np.log(externals/cl)/fix*vmax*cl/(kd+cl)/F #Raimondo -0.117*
    vel1.append(jkcc2)

    jkcc2=51.55*pkcc*(ke*cle-k*cl) #Fraser and Huang 51.55*
    vel2.append(jkcc2)

    jkcc2=0.011125*0.3*(ke*cle-k*cl)/(0.000054*((1+ke*cle/0.000054)*(1+ke/0.009)*(1+cle/0.006)+(1+k*cl/0.000054)*(1+k/0.009)*(1+cl/0.006)))/F
    vel3.append(jkcc2)#Lewin 0.01112*
    
    jkcc2=pkcc*(R*np.log(ke/k)-R*np.log(cl/cle)) #Doyon
    vel0.append(jkcc2)

plt.plot(clspace,vel0,'k')
plt.plot(clspace,vel1,'g')
plt.plot(clspace,vel2,'r')
plt.plot(clspace,vel3,'b')
plt.savefig('clspace.eps')
plt.show()
```
